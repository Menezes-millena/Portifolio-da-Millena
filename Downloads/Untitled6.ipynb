```python
positivos = 0
negativos = 0

while True:
    numero = 0 #int(input("Digite um número inteiro (0 para parar): "))

    if numero == 0:
        break

    if numero > 0:
        positivos += 1
    elif numero < 0:
        negativos += 1

print(f"\nNúmeros positivos: {positivos}")
print(f"Números negativos: {negativos}")

if positivos > negativos:
    print("Há mais números positivos.")
elif negativos > positivos:
    print("Há mais números negativos.")
else:
    print("Há a mesma quantidade de números positivos e negativos.")
```

cada número positivo ou negativo escrito, soma +1 em duas variaveis, e depois verifica qual das duas variaveis é maior, assim dizendo se foram escritos mais números positivos ou negativos

```python
def celsius_para_fahrenheit(celsius):
    return (celsius * 9/5) + 32

def fahrenheit_para_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9

print("Conversor de Temperatura")
unidade = 'C' #input("A temperatura que você vai digitar está em Celsius (C) ou Fahrenheit (F)? ").strip().upper()

if unidade == 'C':
    celsius = 32 #float(input("Digite a temperatura em Celsius: "))
    fahrenheit = celsius_para_fahrenheit(celsius)
    print(f"{celsius:.2f}°C equivale a {fahrenheit:.2f}°F.")
elif unidade == 'F':
    fahrenheit = float(input("Digite a temperatura em Fahrenheit: "))
    celsius = fahrenheit_para_celsius(fahrenheit)
    print(f"{fahrenheit:.2f}°F equivale a {celsius:.2f}°C.")
else:
    print("Unidade inválida. Por favor, digite 'C' para Celsius ou 'F' para Fahrenheit.")

```

O programa ja tem definido as formulas de conversão de celsius para farenheit e vise-versa, então, quando alguém digita uma temperatura em qualquer uma das medidas, ele automaticamentee converte para a outra
